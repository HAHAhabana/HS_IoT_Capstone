```python
#/content로 이동
%cd /

from glob import glob
#"/content/yolov5/JIKIDU-3/train/images/"에 있는 모든 jpg파일의 경로를 train_img_list에 저장
train_img_list = glob('/content/yolov5/JIKIDU-3/train/images/*.jpg')
#"/content/yolov5/JIKIDU-3/valid/images/"에 있는 모든 jpg파일의 경로를 val_img_list에 저장
val_img_list = glob('/content/yolov5/JIKIDU-3/valid/images/*.jpg')

print(len(train_img_list),len(val_img_list))
```

```python
#'train.txt'와 'val.txt'의 내용을 'train_img_list'와 'val_img_list'로 업데이트
with open('/content/yolov5/JIKIDU-3/train.txt','w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/yolov5/JIKIDU-3/val.txt','w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

#data.yaml을 읽기 모드로 오픈하고 내용을 yaml형식으로 파싱하여 data에 저장
with open('/content/yolov5/JIKIDU-3/data.yaml', 'r') as f:
  data = yaml.full_load(f)

print(data)

#data에서 train과 val 부분의 키 값을 각각 'train.txt', 'val.txt'의 값으로 변경
data['train'] = '/content/yolov5/JIKIDU-3/train.txt'
data['val'] = '/content/yolov5/JIKIDU-3/val.txt'

#data.yaml을 쓰기 모드로 오픈하고 data의 변수값을 yaml 형식으로 직렬화해서 data.yaml에 넣는다.
with open('/content/yolov5/JIKIDU-3/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
